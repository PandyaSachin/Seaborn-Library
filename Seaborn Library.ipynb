# Bar plot in Seaborn library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset("penguins")
```

```python
df
```

```python
df["island"]
```

```python
sns.barplot(x=df.island,y=df.bill_length_mm)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df, hue="sex")
plt.show()
```

```python
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm", data=df, hue='sex',order=order1)
plt.show()
```

```python
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm", data=df, hue='sex',order=order1, hue_order=['Female','Male'])
plt.show()
```

```python
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm", data=df,color="orange")
plt.show()
```

```python
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm", data=df, hue='sex',hue_order=['Female','Male'],order=order1, palette="colorblind")
plt.show()
```

```python
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm", data=df, hue='sex',hue_order=['Female','Male'],order=order1, palette="colorblind",
           saturation=1)
plt.show()
```
